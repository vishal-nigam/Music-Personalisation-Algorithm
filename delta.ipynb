# Delta features

Also known as differential and acceleration coefficients. The MFCC feature vector describes only the power spectral envelope of a single frame, but it seems like speech would also have information in the dynamics i.e. what are the trajectories of the MFCC coefficients over time.

input to delta features in MFCC

## use case

It turns out that calculating the MFCC trajectories and appending them to the original feature vector increases ASR performance by quite a bit (if we have 12 MFCC coefficients, we would also get 12 delta coefficients, which would combine to give a feature vector of length 24).

So basically, delta denoted the direction/ trajactory of the MFCC. MFCC must be dynamic in nature, therefore the direction can add extra layer of segregation while distinguishing songs in our project.

Delta-Delta (Acceleration) coefficients are calculated in the same way, but they are calculated from the deltas, not the static coefficients.

## In simpler language: delta is the change in coefficients, and deltaDelta is the change in delta values.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
song = r"C:\Users\Mahip\Documents\Sound recordings\4stroke.m4a"

y ,sr = librosa.load(song)
mfcc = librosa.feature.mfcc(y)
delta = librosa.feature.delta(mfcc)
```

```python
librosa.display.waveplot(delta)
```
