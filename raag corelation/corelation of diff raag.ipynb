```python
def plot_and_corelate(array1 , array2):
    
    corr = signal.correlate(array1 , array2)
    
    fig,axs = plt.subplots(3,1)
    axs[0].plot(array1)
    axs[0].set_ylabel('Original')
    axs[1].plot(array2)
    axs[1].set_ylabel('Mean')
    axs[2].plot(corr)
    axs[2].set_ylabel('correlation')
    
    plt.show()
```
